```python
import fasttext
import sys
```

```python
class SentimentPredictor:
    def __init__(self, model_path):
        """Load the trained FastText model"""
        try:
            self.model = fasttext.load_model(model_path)
            print(f"‚úÖ Model loaded successfully from {model_path}")
            print("üöÄ Ready for predictions! Type your reviews below...")
        except Exception as e:
            print(f"‚ùå Error loading model: {e}")
            sys.exit(1)
    
    def predict_sentiment(self, text):
        """Predict sentiment for a single text"""
        # Clean the text
        clean_text = text.strip().replace('\n', ' ')
        
        # Get prediction
        predictions, confidence_scores = self.model.predict(clean_text, k=1)
        
        # Extract label and confidence
        predicted_label = predictions[0].replace('__label__', '')
        confidence = confidence_scores[0]
        
        return predicted_label, confidence
    
    def predict_with_details(self, text):
        """Get detailed prediction with probabilities for all classes"""
        clean_text = text.strip().replace('\n', ' ')
        
        # Get top 3 predictions to see all probabilities
        predictions, confidence_scores = self.model.predict(clean_text, k=3)
        
        print(f"\nüìä Detailed prediction for: '{text[:50]}...'")
        print("-" * 50)
        
        for i, (pred, conf) in enumerate(zip(predictions, confidence_scores)):
            label = pred.replace('__label__', '')
            print(f"{i+1}. {label}: {conf:.3f}")
        
        # Return the top prediction
        main_label = predictions[0].replace('__label__', '')
        main_confidence = confidence_scores[0]
        
        return main_label, main_confidence
    
    def interactive_mode(self):
        """Run interactive prediction loop"""
        print("\n" + "="*60)
        print("üéØ SENTIMENT ANALYSIS PREDICTION TOOL")
        print("="*60)
        print("Type your review and press Enter to see the prediction")
        print("Type 'quit', 'exit', or press Ctrl+C to stop")
        print("="*60)
        
        while True:
            try:
                # Get input from user
                user_input = input("\nüìù Enter your review: ").strip()
                
                # Check for exit commands
                if user_input.lower() in ['quit', 'exit', 'q', '']:
                    print("üëã Goodbye!")
                    break
                
                if len(user_input) < 3:
                    print("‚ö†Ô∏è  Please enter a longer review (at least 3 characters)")
                    continue
                
                # Get prediction
                predicted_label, confidence = self.predict_sentiment(user_input)
                
                # Display result
                print(f"\nüéØ Prediction: {predicted_label.upper()}")
                print(f"üìà Confidence: {confidence:.3f}")
                
                # Show confidence level
                if confidence > 0.8:
                    confidence_level = "Very confident"
                elif confidence > 0.6:
                    confidence_level = "Confident"
                elif confidence > 0.4:
                    confidence_level = "Moderately confident"
                else:
                    confidence_level = "Not very confident"
                
                print(f"üí° {confidence_level}")
                
            except KeyboardInterrupt:
                print("\n\nüëã Goodbye!")
                break
            except Exception as e:
                print(f"‚ùå Error during prediction: {e}")
```

```python
# Main execution
if __name__ == "__main__":
    # Path to your trained model
    MODEL_PATH = "/home/ilya/Documents/sentiment_model.bin"  # Change this if your file is in a different location
    
    # Create predictor and start interactive mode
    predictor = SentimentPredictor(MODEL_PATH)
    predictor.interactive_mode()
```
